```python
# 1. Load and clean the secondary dataset
def get_wb_data(indicator, country, start_year, end_year):

    url = f"https://api.worldbank.org/v2/country/{country}/indicator/{indicator}?date={start_year}:{end_year}&format=json&per_page=1000"

    response = requests.get(url)

    data = response.json()

    df = pd.DataFrame(data[1])

    return df
```

```python
gdp_df = get_wb_data(
    indicator="NY.GDP.PCAP.CD",
    country="JPN",
    start_year=2000,
    end_year=2023
)

print(gdp_df.head())
```

```python
gdp_clean_df = gdp_df[
    ["countryiso3code", "date", "value"]
]

gdp_clean_df = gdp_clean_df.rename(columns={
    "countryiso3code": "country",
    "date": "year",
    "value": "gdp_per_capita"
})

print(gdp_clean_df.head())
```
